```python



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#
np.random.seed(42)

data = {
    'Month': ['Jan','Feb','Mar','Apr','May','Jun',
              'Jul','Aug','Sep','Oct','Nov','Dec'],
    'Sales': np.random.randint(15000, 80000, 12),
    'Category': np.random.choice(['Electronics','Clothing','Food','Beauty'], 12),
    'Region': np.random.choice(['North','South','East','West'], 12)
}

df = pd.DataFrame(data)
print(df)

```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Sales Analysis Dashboard', fontsize=16, fontweight='bold')

# Chart 1: Monthly Sales
axes[0].bar(df['Month'], df['Sales'], color='steelblue')
axes[0].set_title('Monthly Sales')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Sales ($)')
axes[0].tick_params(axis='x', rotation=45)

# Chart 2: Sales by Category
category_sales = df.groupby('Category')['Sales'].sum()
axes[1].pie(category_sales, labels=category_sales.index, autopct='%1.1f%%')
axes[1].set_title('Sales by Category')

# Chart 3: Sales by Region
region_sales = df.groupby('Region')['Sales'].sum()
axes[2].barh(region_sales.index, region_sales.values, color='coral')
axes[2].set_title('Sales by Region')
axes[2].set_xlabel('Total Sales ($)')

plt.tight_layout()
plt.savefig('sales_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Dashboard saved!")
```

```python
print("=" * 40)
print("📊 SALES ANALYSIS SUMMARY")
print("=" * 40)
print(f"Total Annual Sales: ${df['Sales'].sum():,}")
print(f"Average Monthly Sales: ${df['Sales'].mean():,.0f}")
print(f"Best Month: {df.loc[df['Sales'].idxmax(), 'Month']} (${df['Sales'].max():,})")
print(f"Worst Month: {df.loc[df['Sales'].idxmin(), 'Month']} (${df['Sales'].min():,})")
print()
print("Sales by Category:")
print(df.groupby('Category')['Sales'].sum().sort_values(ascending=False).to_string())
print()
print("Sales by Region:")
print(df.groupby('Region')['Sales'].sum().sort_values(ascending=False).to_string())
```
